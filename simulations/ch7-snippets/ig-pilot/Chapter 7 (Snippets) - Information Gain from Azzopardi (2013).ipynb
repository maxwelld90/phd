```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
mpl.rcParams['pdf.fonttype'] = 42
```

```python
# Read in the summary dataset, compute IG/Word
df = pd.read_csv('sum2.40.txt', header=0, sep=' ')
df = df.drop(columns=['type', 'size', 'smog'])
df['ig_word'] = df.ig / df.words
```

```python
snippets_vs_words = plt.plot(df['snip_en'], df['words'])
```

```python
snippets_vs_ig = plt.plot(df['snip_en'], df['ig'])
```

```python
snippets_vs_ig_word = plt.plot(df['snip_en'], df['ig_word'])
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharex=True, figsize=(10, 3))

ax1.plot(df['snip_en'], df['words'], marker='o', color='b', linewidth=3)
ax2.plot(df['snip_en'], df['ig'], marker='o', color='b', linewidth=3)
ax3.plot(df['snip_en'], df['ig_word'], marker='o', color='b', linewidth=3)

ax1.set_ylabel('Words')
ax1.yaxis.set_ticks(np.arange(0, 121, 20))
ax2.set_ylabel('Information Gain')
ax2.yaxis.set_ticks(np.arange(-4.00, -2.2, 0.25))
ax3.set_ylabel('Information Gain/Word')
ax3.yaxis.set_ticks(np.arange(-0.4, 0.1, 0.1))

f.suptitle('Snippet Fragments: Length and Informativeness', va='bottom')
f.tight_layout()
```

```python
pp = PdfPages('output.pdf')
pp.savefig(f)
pp.close()
```
